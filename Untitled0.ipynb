```python
weights=         [0.30,             0.10,  0.30,        0.10,         0.20  ]
```

```python
pf3=ok.Portfolio(['SP500TR.INDX','AGG.US','MCFTR.INDX','GC.COMM','RGBITR.INDX'],ccy='RUB',weights=weights,rebalancing_period='year')
```

```python
pf3
```

```python
pf3.symbol='rebalanced_portfolio.PF'
```

```python
pf4=ok.Portfolio(['SP500TR.INDX','AGG.US','MCFTR.INDX','GC.COMM','RGBITR.INDX'],ccy='RUB',weights=weights,rebalancing_period='none')
```

```python
pf4.weights_ts.plot()
```

```python
pf3.weights_ts.plot()
```

```python
pf4.symbol='not_rebalanced_portfolio.PF'
```

```python
al=ok.AssetList([pf3,pf4],ccy='RUB')
```

```python
al.wealth_indexes.plot(figsize=[12,6])
```

```python
al.describe([1,5,15])
```
